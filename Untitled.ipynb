```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense,Flatten, Dropout,BatchNormalization, GlobalAveragePooling2D
from keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.optimizers import Adam 
import pandas as pd
import numpy as np 
from matplotlib import pyplot as plt 
import os as os
from sklearn.model_selection import train_test_split
import tensorflow as tf
import cv2 as cv2
from PIL import Image
```

```python
datasets="//content//drive//MyDrive//dataset2"
```

```python
train_path="/content/drive/MyDrive/dataset2/train"
test_path="/content/drive/MyDrive/dataset2/test"
```

```python
def loading_images(path):
    pathnames=[]
    dir_list=os.listdir(path)
    unique_classes=sorted(dir_list)
    for i in range(len(unique_classes)):
        newpath=train_path+"/"+unique_classes[i]
        pathnames.append(newpath)
    return(pathnames,unique_classes)
```

```python
def loading_imagess(path):
    pathnames=[]
    dir_list=os.listdir(path)
    unique_classes=sorted(dir_list)
    for i in range(len(unique_classes)):
        newpath=test_path+"/"+unique_classes[i]
        pathnames.append(newpath)
    return(pathnames,unique_classes)
```

```python
paths,classes=loading_images(train_path)
print(len(paths))
```

```python
testpath,testclass=loading_imagess(test_path)
print(len(testpath))
```

```python
plt.imshow(plt.imread(paths[0]))
plt.imshow(plt.imread(testpath[0]))
```

```python
def labelling_images(path,unique_classes):
    labels=[]
    for label in unique_classes:
        labels.append(label)
    return(labels)
```

```python
labels=labelling_images(paths,classes)
```

```python
testlabels=labelling_images(testpath,testclass)
```

```python
def image_array(paths):
    arr_list=[]
    for i in paths:
        img=cv2.resize(cv2.imread(i),(64,64))
        arr_list.append(img)
    return(arr_list)
    
```

```python
testlabels=np.array(testlabels)
test_img_arr=np.array(testpath)
test_img_arr=np.array(test_img_arr)
```

```python
labels=np.array(labels)
img_arr=image_array(paths)
img_arr=np.array(img_arr)
print(type(img_arr[0]))
plt.imshow(img_arr[0])
```

```python
img=cv2.imread("/content/drive/MyDrive/dataset2/test/2fd875eaa.jpg")
img=cv2.resize(img,(64,64))
plt.imshow(img)
```

```python
X_train,Y_train=(img_arr,labels)
X_test,Y_test=(test_img_arr,testlabels)
train_n=len(X_train)
test_n=len(X_test)
print("Number of training images: " , train_n)
print("Number of testing images: ", test_n)
```

```python
from sklearn.preprocessing import OneHotEncoder
vals = OneHotEncoder()
unq_labes  = np.unique(Y_train)
oneHotLabels = vals.fit_transform(Y_train.reshape(-1,1)).toarray()
Y_test_labs = vals.fit_transform(Y_test.reshape(-1,1)).toarray()
```

```python
print(len(unq_labes))
```

```python
model = Sequential()
model.add(Conv2D(filters = 64, kernel_size = 5, padding = 'same', activation = 'relu',input_shape = (64, 64, 3)))
model.add(Conv2D(filters = 64, kernel_size = 5, padding = 'same', activation = 'relu'))
model.add(MaxPooling2D(pool_size = (4, 4)))
model.add(Dropout(0.5))
model.add(Conv2D(filters = 128 , kernel_size = 5, padding = 'same', activation = 'relu'))
model.add(Conv2D(filters = 128 , kernel_size = 5, padding = 'same', activation = 'relu'))
model.add(MaxPooling2D(pool_size = (4, 4)))
model.add(Dropout(0.5))
model.add(Conv2D(filters = 256 , kernel_size = 5, padding = 'same', activation = 'relu'))
model.add(Dropout(0.5))
model.add(Flatten())
model.add(Dense(3422, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer= Adam(learning_rate=0.01), metrics=['accuracy'])
hist = model.fit(X_train,oneHotLabels,  batch_size =32, validation_data =(X_test,Y_test), epochs=5, verbose=1)
```
